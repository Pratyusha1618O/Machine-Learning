```python
df = pd.read_csv(
    "Market_Basket_Optimisation.csv",
    header=None
)
```

```python
df.head()
```

```python
transactions = []

for i in range(
    0,
    len(df)
):

    transactions.append(
        [
            str(df.values[i,j])
            for j in range(20)
        ]
    )
```

```python
print(transactions[0])
```

```python
rules = apriori(
    transactions,
    min_support=0.003,
    min_confidence=0.2,
    min_lift=3,
    min_length=2
)
```

```python
results = list(rules)

print(
    "Total Rules:",
    len(results)
)
```

```python
results = list(rules)

print(
    "Total Rules:",
    len(results)
)
```

```python
for item in results:

    pair = item[0]

    print(
        list(pair)
    )

    print(
        "Support:",
        item[1]
    )

    print(
        "Confidence:",
        item[2][0][2]
    )

    print(
        "Lift:",
        item[2][0][3]
    )

    print("-"*50)
```

```python
rule_list = []

for item in results:

    rule_list.append([
        list(item[0]),
        item[1],
        item[2][0][2],
        item[2][0][3]
    ])

rules_df = pd.DataFrame(
    rule_list,
    columns=[
        'Rule',
        'Support',
        'Confidence',
        'Lift'
    ]
)

rules_df.head(10)
```

```python
rules_df.sort_values(
    by='Lift',
    ascending=False
).head(10)
```
